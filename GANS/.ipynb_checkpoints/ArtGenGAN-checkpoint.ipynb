## **Function Defs**

```python
def resize_images(inpath, outpath, size):
  fileList = os.listdir(inpath)
  print(f"Resizing {len(fileList)} images")
  for image in fileList:
    if exists(f"{outpath}/{image}"):
        #print("Image exists")
        continue
    try:
        img = cv2.imread(join(inpath, image))
        img = cv2.resize(img, size)
        cv2.imwrite(f"{outpath}/{image}",img)
    except:
        print(f"Error creating {image}")
    # print(img.shape)
```

```python
# Decode a single image in a TFRecord
# It also standardizes the image
def decode_image(image) -> tf.Tensor:
    #img = (tf.cast(image, tf.float32) / 127.5) - 1
    img = image / 127.5 - 1
    print(type(img))
    img = tf.reshape(img, [img_width, img_height, channels])
    #print(img.shape)
    return img
```

```python
def dataset_from_dir(dataset_folder, ordered=False):
    dataset = tf.keras.utils.image_dataset_from_directory(
        dataset_folder,
        labels=None,
        #validation_split=0.2,
        #subset="training",
        shuffle = False,
        seed=seed,
        image_size=(img_height, img_width),
        batch_size=1)
    dataset = dataset.map(decode_image)
    dataset = dataset.cache()
    #dataset.shuffle(2048)
    return dataset.batch(batch_size)
```

```python
# Memory intensive. Use load_batch instead. 
def dataset_manual(dataset_folder, max=0):
    imageNames = os.listdir(resized_path)
    if max > 0:
        imageNames = imageNames[:max]
        print(f"using {len(imageNames)} images.")
    array = []
    for image in imageNames:
        image = Image.open(join(resized_path, image))
        data = np.asarray(image)
        array.append(data)

    X_train = np.array(array)
    X_train = X_train / 127.5 -1.
    print(X_train.shape)
    return X_train
```

```python
def get_images_list(path):
    files = []
    # r=root, d=directories, f = files
    for r, d, f in os.walk(path):
        for file in f:
            if '.jpg' in file or '.png' in file:
                files.append(os.path.join(r, file))
    return files
```

```python
# Colab was running out of memory because the datasets were huge. This solves that. 
def load_batch(fileList, batchSize):
    array = []

    for i in range(batchSize):
        # if there is less than 3 color channels, keep trying. 
        while True:
            file = random.choice(fileList)
            image = Image.open(file)
            if len(image.split()) != 3:
                print(f"removing {file}")
                os.remove(file)
            else:
                break
        data = np.asarray(image)
        data = data / 127.5 -1.
        #print(data)
        array.append(data)

    train = np.array(array, dtype=np.float32)
    return train
```

```python
def save_image(name, source):
    image = Image.fromarray(np.uint8(source*255))
    image.save(name)
```

## **3) Building Generator**





```python
def build_generator():
    multiplyer = int(((img_width/2)/2/2))
    model = Sequential(name="Generator")
    model.add(Dense(img_width * multiplyer * multiplyer, input_dim=latent_dim)) # img_width was 256 here. 
    model.add(LeakyReLU(alpha=0.3))
    model.add(Reshape((multiplyer,multiplyer,img_width))) # img_width was 256 here. 
    
    #model.add(BatchNormalization()) # Added
    
    model.add(Conv2DTranspose(128, (4,4), strides=(1,1), padding='same'))
    model.add(LeakyReLU(alpha=0.3)) # activation leakey instead of old fashion RelU

    model.add(Conv2DTranspose(128, (4,4), strides=(2,2), padding='same'))
    model.add(LeakyReLU(alpha=0.2))

    model.add(Conv2DTranspose(128, (2,2), strides=(2,2), padding='same'))
    model.add(LeakyReLU(alpha=0.3))
    
    model.add(Conv2DTranspose(128, (4,4), strides=(2,2), padding='same'))
    model.add(LeakyReLU(alpha=0.3))

    model.add(Conv2D(3, (3,3), activation='tanh', padding='same'))
  
    model.summary()

    return model

generator = build_generator()
```

## **4) Building Discriminator**

```python
def build_discriminator():
    model = Sequential(name="Discriminator")
    model.add(Conv2D(64, (3,3), padding='same', input_shape=img_shape))
    model.add(LeakyReLU(alpha=0.3))

    model.add(Conv2D(128, (3,3), padding='same', ))
    model.add(LeakyReLU(alpha=0.3))
    
    model.add(Conv2D(128, (3,3), padding='same'))
    model.add(LeakyReLU(alpha=0.3))

    model.add(Conv2D(filters=256, kernel_size=(2,2), padding='same'))
    model.add(LeakyReLU(alpha=0.3))
    
    model.add(Conv2D(256, (3,3), padding='same'))
    model.add(LeakyReLU(alpha=0.3))

    model.add(Flatten())
    model.add(Dropout(0.4))
    model.add(Dense(1, activation='sigmoid'))

    model.summary()
    return model

discriminator = build_discriminator()
discriminator.compile(loss='binary_crossentropy', optimizer=adam, metrics=['accuracy'])

discriminator.trainable = False # Set to false before compiling main GAN.
```

## **5) Connecting Neural Networks to build GAN**

```python
GAN = Sequential(name="THE_GAN")
GAN.add(generator)
GAN.add(discriminator)
GAN.summary()
GAN.compile(loss='binary_crossentropy', optimizer=adam)
```

## **6) Outputting Images**


```python
import matplotlib.pyplot as plt
import glob
import imageio
import PIL

def save_imgs(epoch):
    generator, discriminator = GAN.layers
    r, c = 4, 4
    #noise = np.random.normal(0, 1, (r * c, latent_dim))
    noise = tf.random.normal(shape=[r * c, latent_dim])
    gen_imgs = generator(noise)

    # Rescale images 0 - 1
    gen_imgs = (gen_imgs + 1) / 2.0
    #gen_imgs = (gen_imgs * 255.).astype(int)
    #gen_imgs = np.clip(gen_imgs, 0, 255)
    #print(f"{gen_imgs.shape}\n{gen_imgs[0]}")
    fig, axs = plt.subplots(r, c)
    cnt = 0
    for i in range(r):
        for j in range(c):
            axs[i,j].imshow(gen_imgs[cnt])
            axs[i,j].axis('off')
            save_image(f"{generated_path}/{epoch}_{gan_name}_{cnt}.jpg",gen_imgs[cnt])
            cnt += 1
    print(f"Saving Image to {generated_path}")
    fig.savefig("currentgeneration.png")
    #fig.savefig(f"{generated_path}/{epoch}_{gan_name}_plot.png")
    plt.close()

#save_imgs(0)
```

## **7) Training GAN**

```python
def train(epochs, batch_size=32):
    # Grab the seprate components
    generator, discriminator = GAN.layers
    bat_per_epo = epoch_batches

    print(f"training starting with {epochs} epochs of {epoch_batches} total batches each. Batch size is {batch_size}")
    #train = np.expand_dims(X_train, axis=3)
    #print(f"Train shape = {train.shape}")

    #Create our Y for our Neural Networks
    valid = np.ones((batch_size, 1))
    fakes = np.zeros((batch_size, 1))

    if exists(f"{model_save_path}/{gan_name}_generator.h5"):
        print("Loading generator weights.")
        generator.load_weights(f"{model_save_path}/{gan_name}_generator.h5")
    if exists(f"{model_save_path}/{gan_name}_generator.h5"):
        print("Loading discriminator weights.")
        discriminator.load_weights(f"{model_save_path}/{gan_name}_discriminator.h5")

    for epoch in range(epochs):
        for j in range(bat_per_epo):
            #Get Random Batch
            imgs = load_batch(imageFiles, batch_size)
            
            #Phase 1: Generate fake images.
            #noise = np.random.normal(0, 1, (batch_size, latent_dim))
            noise = tf.random.normal(shape=[batch_size, latent_dim])
            #gen_imgs = generator.predict(noise)
            gen_imgs = generator(noise)

            #Phase 2: Train discriminator
            discriminator.trainable = True # I think we forgot this so it never trained, ever. 
            d_loss_fake = discriminator.train_on_batch(gen_imgs, fakes)
            d_loss_real = discriminator.train_on_batch(imgs, valid)

            d_loss = 0.5 * np.add(d_loss_real, d_loss_fake)

            # Phase 2: feed fake images to generator and set labels as "real". 
            # Avois a warning
            discriminator.trainable = False
            
            #noise = np.random.normal(0, 1, (batch_size, latent_dim))
            noise = tf.random.normal(shape=[batch_size, latent_dim])
            
            g_loss = GAN.train_on_batch(noise, valid)
            if j % 100 == 0:
              print(f"******* Epoch: {epoch+1}/{epochs}: Batch {j}/{bat_per_epo} [D loss: {d_loss[0]}, acc: {100* d_loss[1]:.4f}] [G loss: {g_loss}]")

        # if(epoch % save_interval) == 0:
        generator.save_weights(f"{model_save_path}/{gan_name}_generator.h5")
        discriminator.save_weights(f"{model_save_path}/{gan_name}_discriminator.h5")
        save_imgs(epoch)
    print("Training Done!")
```

```python
startover = False
if startover:
    print("Removing saved models.")
    os.remove(f"{model_save_path}/{gan_name}_generator.h5")
    os.remove(f"{model_save_path}/{gan_name}_discriminator.h5")
```
